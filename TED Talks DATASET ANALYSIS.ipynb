#                                  TED Talks DATASET ANALYSIS

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('ted_main.csv')
```

# Maximum Performers at TED Talks

```python
df1 = df.iloc[0:200]
```

```python
d1=df1['event'].value_counts()
d1.plot.bar()
```

# Main speakers of TED Talks with more than one performer

```python
df2 = df.iloc[0:500]
```

```python
d2 = df2[df2['num_speaker']>1]['main_speaker']
print(d2)
```

#  Languages vs Duration of TED Talks

```python
d3 = df1.plot(kind='scatter',x='languages',y='duration')
d3
```

# Views vs Comments of TED Talks

```python
df3 = df.iloc[250:260]
d4 = df3.plot(kind='line',x='views',y='comments',color='red')
d4
```

# Total comments of each TED Talks event

```python
df1.groupby('event')['comments'].nunique().plot(kind='bar')
plt.show()
```

# Event and No. of speakers

```python
d2 = pd.crosstab(df1['event'],df1['num_speaker'],margins=True) 
d2
```

```python
d2.plot.bar(figsize = (10,8),ylim= [0,50])
```

# Duration and Title of the TED Talks

```python
d2 = df1[['duration','title']]
print(d2)
```

# Published_Date and Duration of TED talks of event TED2006 

```python
d2 = df1[(df1['event']=='TED2006') & (df1['duration']>1300)]['published_date'].value_counts()
d2.plot.pie()
```

# Events and Published_dates

```python
d2 = pd.crosstab(df1['event'],df1['published_date'],margins=True) 
d2
```

# Language and Comments of TED Talks of event TED2007

```python
d4 = df2[(df2['event']=='TED2007') & (df2['comments']>100)]['languages'].value_counts()
d4
```

```python
d4.plot()
```
